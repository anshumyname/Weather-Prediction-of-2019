#                                     PREDICTING WEATHER OF 2019                                       #

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= pd.read_excel("imported1.xlsx")

df.head()
```

```python
test= df[' _tempm']
test.fillna(test.mean(), inplace=True)
```

```python
train= df.drop([' _precipm',' _wgustm',' _windchillm',' _wdird',' _heatindexm',' _fog', ' _hail',' _rain',' _snow',' _thunder',' _tornado', ' _wdire',' _tempm'], axis=1)
train.fillna(train.mean(), inplace=True)
```

```python
from sklearn.cross_validation import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(train,test,test_size = 0.25,random_state = 101)
x_train
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
prediction = reg.predict(x_test)
```

```python
plt.scatter(prediction,y_test)
plt.ylabel('ACTUAL VALUE')
plt.xlabel('PREDICTIONS')
```

```python
reg.score(x_test,y_test)
```

```python
X= reg.coef_
X
```
